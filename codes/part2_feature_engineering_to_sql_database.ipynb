## Indivual Hitters Page - Final

```python
#2015-2019 player list (>162 PA on one single season)
players=pd.read_csv('csv_year/mlb_batting_2019.csv')[['Html_id','Name','Year','PA']]
for season in range(2018,2014,-1):
    players = players.append(pd.read_csv(f'csv_year/mlb_batting_{season}.csv')[['Html_id','Name','Year','PA']])
players = players[(players['PA'] >= 162) & (players['Html_id'] != 'None')]
players = players.groupby(['Html_id','Name'], as_index=False).PA.sum()
players.sort_values('PA', ascending = False, inplace= True)
players.rename(columns={"PA": "2015_2019_PA"}, inplace=True)
players.reset_index(drop=True, inplace=True)
players.to_csv('hitters_2015_2019.csv')
#players.drop_duplicates("Html_id", inplace=True)
```

```python
#########must run#########
def feature_add(df_player):
    '''
    add some features
    '''
    df_player['PA'].fillna(0, inplace=True)
    #raw_data['OPS+'].fillna(0, inplace=True)
    df_player['PA_cum'] = df_player['PA'].astype(int).cumsum(axis=0)
    df_player['PA_cum_0'] = df_player['PA_cum'].astype(int) - df_player['PA'].astype(int)
    df_player['MLB_year'] = df_player.index
    df_player["PA_0"] = df_player.PA.shift()
    df_player["PA_3y"] = df_player.PA.rolling(3, min_periods=1).mean() # average PA of previous 3 years
    df_player["PA_3y_0"] = df_player.PA_3y.shift()
    df_player["PA_2y"] = df_player.PA.rolling(2, min_periods=1).mean() # average PA of previous 2 years
    df_player["PA_2y_0"] = df_player.PA_2y.shift()
    df_player["OPS+_3y"] = df_player["OPS+"].rolling(3, min_periods=1).mean() # average OPS+ of previous 3 years
    df_player["OPS+_3y_0"] = df_player["OPS+_3y"].shift()
    df_player["OPS+_2y"] = df_player["OPS+"].rolling(2, min_periods=1).mean() # average OPS+ of previous 2 years
    df_player["OPS+_2y_0"] = df_player["OPS+_2y"].shift()
    df_player["OPS+_last"] = df_player["OPS+"].shift()
    df_player["OPS+_next"] = df_player["OPS+"].shift(-1)
    return df_player
```

```python
#########must run#########
#combine player data
df = pd.read_csv('csv_player/pierza.01.csv')
feature_add(df)
for p in range(0, 416):
    df_con = pd.read_csv('csv_player/' + players['Html_id'][p].split('.')[0].split('/')[3] + '.csv')
    feature_add(df_con)
    df = df.append(df_con)
    print(str(p) + ' done!')

for p in range(417, 681):
    df_con = pd.read_csv('csv_player/' + players['Html_id'][p].split('.')[0].split('/')[3] + '.csv')
    feature_add(df_con)
    df = df.append(df_con)
    print(str(p) + ' done!')
```
